```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Load dataset
iris_df = sns.load_dataset("iris")
tips_df = sns.load_dataset("tips")
titanic_df = sns.load_dataset("titanic")
flights_df = sns.load_dataset("flights")
```

```python
# tips = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
# iris = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
```

### Univariate Analysis

#### Histogram
- Understand the distribution of a single numerical variable.
- Visualize the distribution of sepal length in the Iris dataset.

```python
plt.hist(iris_df['sepal_length'], bins=15, edgecolor='black')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length')
plt.show()
```

#### Bar plot
- Compare categorical data across different groups.
- Represent the count of passengers in different classes in the Titanic dataset.

```python
class_counts = titanic_df['class'].value_counts()
plt.bar(class_counts.index, class_counts.values)
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.title('Bar Plot of Passenger Class Distribution')
plt.show()
```

#### Count Plot
- Visualize the count of observations in each category.
- Show the count of different species in the Iris dataset.

```python
sns.countplot(data=iris_df, x='species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Count Plot of Species Distribution')
plt.show()
```

#### Box plot
- Display the distribution of data through quartiles and outliers.
- Visualize the distribution of sepal lengths in the Iris dataset.

```python
plt.boxplot(iris_df['sepal_length'])
plt.ylabel('Sepal Length')
plt.title('Box Plot of Sepal Length')
plt.show()
```

#### Violin plot
- Combine a box plot with a rotated kernel density plot.
- Combine box plot and kernel density estimation for the distribution of sepal lengths.

```python
plt.violinplot(iris_df['sepal_length'], vert=False)
plt.xlabel('Sepal Length')
plt.title('Violin Plot of Sepal Length')
plt.show()
```

#### Distribution Plot
- Display the distribution of a single variable.
- Visualize the distribution of tip amounts in the Tips dataset.

```python
sns.distplot(tips_df['tip'], hist=True, kde=True, bins=15)
plt.xlabel('Tip Amount')
plt.ylabel('Density')
plt.title('Distribution Plot of Tip Amount')
plt.show()
```

#### Line Plot
- It shows the distribution of a single variable (usually over time) by connecting data points with lines.

```python
plt.plot(flights_df['year'], flights_df['passengers'])
plt.title("Line Plot of Passengers Over Years")
plt.xlabel("Year")
plt.ylabel("Passengers")
```

#### Strip Plot
-  It displays individual data points along a categorical axis, helping to visualize the distribution of a single variable.

```python
plt.scatter(iris_df['species'], iris_df['petal_length'])
plt.title("Strip Plot of Petal Length for Different Species")
plt.xlabel("Species")
plt.ylabel("Petal Length")
plt.xticks(rotation=45)
```

### Bivariate Analysis

#### Pair plot
- Visualize relationships between multiple numerical variables.
- Visualize pairwise relationships among numerical columns in the Iris dataset.

```python
sns.pairplot(iris_df, hue='species', diag_kind='hist')
plt.title('Pair Plot of Numerical Variables in Iris Dataset')
plt.show()
```

#### Correlation Heatmap
- Understand correlations between numerical variables.
- Visualize correlations between numerical variables in the Titanic dataset.- 

```python
correlation_matrix = titanic_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Titanic Dataset')
plt.show()
```

#### Scatter Plot
- Visualize the relationship between two continuous variables
- Visualize the relationship between sepal length and sepal width in the Iris dataset.

```python
plt.scatter(iris_df['sepal_length'], iris_df['sepal_width'], c=iris_df['species'].astype('category').cat.codes, cmap='viridis')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.show()
```

#### Joint Plot
- Combine multiple plots for two variables, like scatter plots and histograms.
- Combine scatter plot and histograms for the relationship between total bill and tip.

```python
sns.jointplot(x='total_bill', y='tip', data=tips_df, kind='scatter')
plt.title('Joint Plot of Total Bill vs Tip')
plt.show()
```

#### Regression Plot (lmplot)
- Visualize the linear relationship between two variables.
-  It shows the relationship between two continuous variables by fitting a regression line.
- Visualize the linear relationship between total bill and tip in the Tips dataset.

```python
sns.regplot(x='total_bill', y='tip', data=tips_df)
plt.xlabel('Total Bill')
plt.ylabel('Tip Amount')
plt.title('Regression Plot of Total Bill vs Tip')
plt.show()
```

#### Categorical Plot
- Show relationships between a categorical and a continuous variable.
- Show the relationship between day of the week and total bill using a bar plot.

```python
sns.catplot(x='day', y='total_bill', data=tips_df, kind='bar', palette='Set2')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill')
plt.title('Total Bill by Day of the Week')
plt.show()
```

#### Swarm Plot
- It displays individual data points for two variables, allowing you to see the distribution of one variable across categories of another variable.

```python
plt.scatter(titanic_df['class'], titanic_df['age'])
plt.title("Swarm Plot of Age by Passenger Class")
plt.xlabel("Class")
plt.ylabel("Age")
```

### Multivariate Analysis

#### Pair grid
- Create subplots for pairwise relationships in a dataset.
- Create subplots for pairwise relationships among numerical columns in the Tips dataset.

```python
g = sns.PairGrid(tips_df)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
plt.suptitle('Pair Grid of Numerical Variables in Tips Dataset')
plt.show
```

#### Facet Grid
- Create a grid of subplots based on different categories.
- Create subplots based on different days of the week to compare total bills.

```python
# Facet Grid: Create subplots based on different days of the week to compare total bills.
g = sns.FacetGrid(tips_df, col='day')
g.map(sns.histplot, 'total_bill')
plt.suptitle('Facet Grid of Total Bill Distribution by Day of the Week', y=1.02)  # Added spacing
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

#### Time Series Plot
- Visualize data points over time, crucial for time series analysis.
- Visualize the number of flights over the years.

```python
flights_pivot = flights_df.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(flights_pivot, cmap='YlGnBu', annot=True, fmt='d')
plt.title('Time Series Heatmap of Flights')
plt.show()
```
