Univariate Analysis:

- Histogram: Distribution of a single variable.
- Bar Plot: Categorical data representation.
- Count Plot: Count of observations in each category.
- Box Plot: Distribution of data through quartiles and outliers.
- Violin Plot: Combines box plot with kernel density plot for distribution visualization.
- Distribution Plot: Visualizes the distribution of a single variable.
- Line Plot: Shows the distribution of a single variable (usually over time) by connecting data points with lines.
- Strip Plot: Displays individual data points along a categorical axis, helping to visualize the distribution of a single variable.

Bivariate Analysis:

- Pair Plot: Pairwise relationships between multiple variables.
- Heatmap: Visualizes relationships and patterns in a matrix of data.
- Correlation Heatmap: Shows correlations between variables using color-coded values.
- Scatter Plot: Visualizes relationships between two continuous variables.
- Joint Plot: Combines multiple plots for two variables (scatter plots and histograms).
- Regression Plot (lmplot): Visualizes the linear relationship between two variables.
- Categorical Plot: Shows relationships between categorical and continuous variables.
- Swarm Plot: Bivariate - It displays individual data points for two variables, allowing you to see the distribution of one variable across categories of another variable.


Multivariate Analysis:

- Pair Grid: Creates subplots for pairwise relationships in a dataset.
- Facet Grid: Creates subplots based on different categories, allowing comparison.
- Time Series Plot: Visualizes data points over time, often crucial for time series analysis.

Other Libraries' Plots:

- Network Graph (using NetworkX library)
- Geospatial Map Plotting (using Geopandas, Folium, etc.)
- Word Cloud (using WordCloud library)

### Univariate Analysis

#### Histogram
- Understand the distribution of a single numerical variable.

```python
# Create histogram
sns.histplot(data=tips_df, x="total_bill");
```

#### Bar plot
- Compare categorical data across different groups.

```python
# Create bar plot
sns.barplot(data=titanic_df, x="class", y="fare")
```

#### Count Plot
- Visualize the count of observations in each category.

```python
# Create count plot
sns.countplot(data=titanic_df, x="class")
```

```python
tips_df.describe()
```

#### Box plot
- Display the distribution of data through quartiles and outliers.

```python
# Create box plot
sns.boxplot(data=tips_df, x="day", y="total_bill")
```

#### Violin plot
- Combine a box plot with a rotated kernel density plot.

```python
# Create violin plot
sns.violinplot(data=tips_df, x="day", y="total_bill")
```

#### Distribution Plot
- Display the distribution of a single variable.

```python
# Create distribution plot
sns.displot(data=tips_df, x="total_bill", kde=True)
```

#### Line Plot
-  It shows the distribution of a single variable (usually over time) by connecting data points with lines.

```python
sns.lineplot(data=flights_df, x="year", y="passengers")
plt.title("Line Plot of Passengers Over Years")
```

### Bivariate Analysis

#### Pair plot
- Visualize relationships between multiple numerical variables.

```python
# Create pair plot
sns.pairplot(iris_df);
```

```python
# Create pair plot
sns.pairplot(iris_df, hue="species");
```

#### Heat map
- Visualize data in a matrix form with color gradients.

```python
pivot_data = flights_df.pivot_table(index="month", columns="year", values="passengers")
pivot_data
```

```python
# Create heatmap
sns.heatmap(data=pivot_data, cmap="YlGnBu")
```

#### Correlation Heatmap
- Understand correlations between numerical variables.

```python
# Calculate correlation matrix
corr_matrix = iris_df.corr()
corr_matrix
```

```python
# Create correlation heatmap
sns.heatmap(data=corr_matrix, annot=True, cmap="coolwarm")
```

#### Scatter Plot
- Visualize the relationship between two continuous variables

```python
# Create scatter plot
sns.scatterplot(data=tips_df, x="total_bill", y="tip")
```

#### Joint Plot
- Combine multiple plots for two variables, like scatter plots and histograms.

```python
# Create joint plot
sns.jointplot(data=tips_df, x="total_bill", y="tip")
```

#### Regression Plot (lmplot)
- Visualize the linear relationship between two variables.
-  It shows the relationship between two continuous variables by fitting a regression line.

```python
# Create regression plot
sns.regplot(data=tips_df, x="total_bill", y="tip")
```

#### Categorical Plot
- Show relationships between a categorical and a continuous variable.

```python
# Create categorical plot
sns.catplot(data=tips_df, x="day", y="total_bill", kind="bar")
```

#### Swarm Plot
-  It displays individual data points for two variables, allowing you to see the distribution of one variable across categories of another variable.

```python
sns.swarmplot(data=titanic_df, x="class", y="age")
plt.title("Swarm Plot of Age by Passenger Class")
```

### Multivariate Analysis

#### Pair grid
- Create subplots for pairwise relationships in a dataset.

```python
# Create pair grid
g = sns.PairGrid(iris_df, hue="species")
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.histplot)
```

#### Facet Grid
- Create a grid of subplots based on different categories.

```python
# Create facet grid
g = sns.FacetGrid(tips_df, col="time", row="sex")
g.map_dataframe(sns.scatterplot, x="total_bill", y="tip")
```

#### Time Series Plot
- Visualize data points over time, crucial for time series analysis.

```python
# Create time series plot
sns.lineplot(data=flights_df, x="year", y="passengers")
```

#### Network Graph (using NetworkX library):
- Visualizing relationships between nodes and edges in a network.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()
G.add_edge("Alice", "Bob")
G.add_edge("Bob", "Charlie")
G.add_edge("Charlie", "David")
G.add_edge("Bob", "David")

# Draw the graph
nx.draw(G, with_labels=True, font_weight='bold')
plt.show()
```

#### Word Cloud (using WordCloud library):
- Visualizing word frequency in text data.

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Sample text
text = "word cloud example for text visualization"

# Create a WordCloud object
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display the word cloud
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import nltk
from nltk.corpus import movie_reviews
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Download the movie_reviews dataset from NLTK
nltk.download('movie_reviews')
```

```python
# Get the movie_reviews dataset
reviews = movie_reviews.raw()

# Create a WordCloud object
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(reviews)

# Display the word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
