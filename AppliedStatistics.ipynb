Central Limit Theorem:
- magic rule
- for large n (each sample size), sampling distribution means -> normally distributed, no matter what the population distribution is.

Assumptions:
- Choose Random sampled data
- Choose Samples from same distribution
- Choose Independent samples
- Sample size must be sufficiently large (>=30)
    - mean of sample means get closer to the population mean  
      
  - Practical understanding of Central Limit Theorem
    - 1000 students in class
    - ht of 1000 students -> any distribution
    - take 10 students as one sample, and choose >=30 samples of that kind
    - find average ht from each sample
    - all these X-bar will form normal distribution: Central limit Theorem
    - From this we can guess: How many students might be a certain height in the population of 1000 students

```python
# Bernoulli Distribution
p_success = 0.7
bernoulli_data = np.random.choice([0, 1], size=100, p=[1 - p_success, p_success])

# Binomial Distribution
num_trials = 10
p_success = 0.5
k = np.arange(0, num_trials + 1)
binom_pmf = stats.binom.pmf(k, num_trials, p_success)
binom_cdf = stats.binom.cdf(k, num_trials, p_success)

# Probability of 5 heads
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
sns.barplot(x=k, y=binom_pmf)
plt.title("Probability of Getting 5 Heads (Binomial PMF)")

# Probability of getting at most 3 heads (<=3)
plt.subplot(1, 2, 2)
sns.lineplot(x=k, y=binom_cdf)
plt.title("Probability of Getting at Most 3 Heads (Binomial CDF)")
plt.tight_layout()
plt.show()

# Poisson Distribution
average_rate = 20
num_samples = 1000
x = np.arange(0, 50)
poisson_pmf = stats.poisson.pmf(x, average_rate)
poisson_cdf = stats.poisson.cdf(x, average_rate)

# Probability of receiving exactly 25 calls in an hour
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.bar(x, poisson_pmf)
plt.title("Probability of Receiving Exactly 25 Calls (Poisson PMF)")

# Probability of receiving 30 or fewer calls in an hour
plt.subplot(1, 3, 2)
plt.step(x, poisson_cdf)
plt.title("Probability of Receiving 30 or Fewer Calls (Poisson CDF)")
plt.axvline(x=30, color='r', linestyle='--', label="Threshold: 30 calls")
plt.legend()

# Probability of receiving more than 30 calls in an hour
plt.subplot(1, 3, 3)
plt.step(x, 1 - poisson_cdf)
plt.title("Probability of Receiving More Than 30 Calls (Poisson CDF)")
plt.axvline(x=30, color='r', linestyle='--', label="Threshold: 30 calls")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# central limit theorem
# Parameters
population_size = 1000
sample_sizes = [2, 30, 400, 800]
num_samples = 1000

# Define distributions
distributions = {
    "Exponential": np.random.exponential(scale=2, size=population_size),
    "Uniform": np.random.uniform(low=0, high=10, size=population_size),
    "Binomial": np.random.binomial(n=20, p=0.5, size=population_size)
}

# Generate and plot sample means and original distribution for each distribution and sample size
for dist_name, population_data in distributions.items():
    plt.figure(figsize=(15, 8))
    plt.suptitle(f"Central Limit Theorem for {dist_name} Distribution")

    # Plot original distribution
    plt.subplot(2, len(sample_sizes) + 1, 1)
    sns.histplot(population_data, bins=50, kde=True, color='green')
    plt.title("Original Population Distribution")
    plt.xlabel("Value")
    plt.ylabel("Density")

    for i, sample_size in enumerate(sample_sizes):
        sample_means = []

        for _ in range(num_samples):
            sample = np.random.choice(population_data, size=sample_size, replace=False)
            sample_mean = np.mean(sample)
            sample_means.append(sample_mean)

        # Plot sample means distribution
        plt.subplot(2, len(sample_sizes) + 1, i + 2)
        sns.histplot(sample_means, bins=30, kde=True, color='blue')
        plt.title(f"Sample Size: {sample_size}")
        plt.xlabel("Sample Mean")
        plt.ylabel("Density")

    plt.tight_layout()
    plt.show()
```

Hypothesis Testing:

- Null Hypothesis (H0)
    - Ex: The new drug has no effect on patient recovery time.
- Alternative Hypothesis (H1)
    - Ex: The new drug significantly reduces patient recovery time.
- Type I Error (False Positive)
    - Ex: Rejecting H0 when it's actually true.
- Type II Error (False Negative)
    - Ex: Failing to reject H0 when it's actually false.
- Significance Level ()
    - Ex: Set at 0.05 for a 5% chance of Type I error.
- p-value
    - Ex: p < 0.05 indicates significant results.
    - Significance Level (): Set at 0.05 for a 5% chance of Type I error

```python
# Sample data (recovery times in hours)
new_drug_recovery = [4, 5, 3, 6, 4, 5]
old_drug_recovery = [6, 7, 5, 8, 6, 7]

# Hypotheses
H0 = "The new drug has no effect on patient recovery time."
H1 = "The new drug significantly reduces patient recovery time."

# Significance level
alpha = 0.05

# Two-Sample t-test (Independent Samples)
t_stat, p_value = stats.ttest_ind(new_drug_recovery, old_drug_recovery)

# Interpretation
print("Null Hypothesis (H0):", H0)
print("Alternative Hypothesis (H1):", H1)
print("Significance Level ():", alpha)
print("p-value:", p_value)

# Check if p-value is less than alpha to decide whether to reject H0
if p_value < alpha:
    print("Reject the Null Hypothesis (H0). There's a significant difference.")
else:
    print("Fail to reject the Null Hypothesis (H0).")
```

- Z-test
    - Ex: Comparing a sample mean with a population mean using z-scores.
    - H0: The average height of male students is equal to the national average height.
    - H1: The average height of male students is different from the national average height.

```python
# Sample data (heights of male students)
sample_heights = np.array([175, 180, 178, 182, 177, 179, 176, 181, 183, 185])

# Population parameters (national average height and standard deviation)
population_mean = 170  # National average height
population_stddev = 8  # Population standard deviation

# Hypotheses
H0 = "The average height of male students is equal to the national average height."
H1 = "The average height of male students is different from the national average height."

# Significance level
alpha = 0.05

# Calculate sample statistics
sample_mean = np.mean(sample_heights)
sample_size = len(sample_heights)

# Calculate z-score
z_score = (sample_mean - population_mean) / (population_stddev / np.sqrt(sample_size))

# Calculate p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(np.abs(z_score)))

# Interpretation
print("Null Hypothesis (H0):", H0)
print("Alternative Hypothesis (H1):", H1)
print("Significance Level ():", alpha)
print("p-value:", p_value)

# Check if p-value is less than alpha to decide whether to reject H0
if p_value < alpha:
    print("Reject the Null Hypothesis (H0). The average height is different.")
else:
    print("Fail to reject the Null Hypothesis (H0). The average height is equal.")
```

- t-test
    - Ex: Comparing means of two samples to test the difference.
    - One-Tailed t-test
        - Ex: Testing if a new training method increases employees' productivity (H1: 1 > 2).
    - Two-Tailed t-test
        - Ex: Testing if there's a difference in the scores of students using two teaching methods (H1: 1  2).
    - One-Sample t-test
        - Ex: Testing if the average response time of a website is significantly less than 3 seconds.
    - Two-Sample t-test (Independent Samples)
        - Ex: Testing if the effectiveness of two different advertising campaigns is different.
    - Paired Sample t-test
        - Ex: Testing if there's a significant reduction in patients' cholesterol levels after a new treatment.

```python
# Sample data
before_training = np.array([60, 65, 58, 63, 61])
after_training = np.array([68, 72, 65, 70, 69])

campaign1 = np.array([23, 27, 30, 28, 25])
campaign2 = np.array([20, 24, 29, 26, 22])

before_treatment = np.array([200, 210, 195, 205, 198])
after_treatment = np.array([180, 185, 175, 190, 182])

# Significance level
alpha = 0.05

# One-Tailed t-test
t_stat1, p_value1 = stats.ttest_rel(after_training, before_training, alternative='greater')
hypothesis1 = "H0: The new training method has no effect on employees' productivity."
hypothesis1_alternative = "H1: The new training method increases employees' productivity."

# Two-Tailed t-test (Independent Samples)
t_stat2, p_value2 = stats.ttest_ind(campaign1, campaign2)
hypothesis2 = "H0: There's no difference in the effectiveness of the two advertising campaigns."
hypothesis2_alternative = "H1: There's a difference in the effectiveness of the two advertising campaigns."

# Paired Sample t-test
t_stat3, p_value3 = stats.ttest_rel(after_treatment, before_treatment)
hypothesis3 = "H0: There's no significant reduction in patients' cholesterol levels after the new treatment."
hypothesis3_alternative = "H1: There's a significant reduction in patients' cholesterol levels after the new treatment."

# One-Sample t-test
sample_data = np.array([28.5, 31.2, 29.8, 30.5, 29.0])
t_stat4, p_value4 = stats.ttest_1samp(sample_data, popmean=30)
hypothesis4 = "H0: The average response time of the website is equal to 30 seconds."
hypothesis4_alternative = "H1: The average response time of the website is different from 30 seconds."

# Two-Sample t-test (Independent Samples)
t_stat5, p_value5 = stats.ttest_ind(campaign1, campaign2)
hypothesis5 = "H0: There's no difference in the effectiveness of the two advertising campaigns."
hypothesis5_alternative = "H1: There's a difference in the effectiveness of the two advertising campaigns."

# Interpretation
print("- t-test")
print("  - Ex: Comparing means of two samples to test the difference.")

# One-Tailed t-test
print("  - One-Tailed t-test")
print("    - Ex: Testing if a new training method increases employees' productivity.")
print("    - H0: No effect on productivity.")
print("    - H1: Increases productivity.")
print("    - p-value:", p_value1)
print("    - Result:", "Reject H0" if p_value1 < alpha else "Fail to reject H0")

# Two-Tailed t-test (Independent Samples)
print("  - Two-Tailed t-test (Independent Samples)")
print("    - Ex: Testing if there's a difference in student scores using two teaching methods.")
print("    - H0: No difference in effectiveness.")
print("    - H1: Difference in effectiveness.")
print("    - p-value:", p_value2)
print("    - Result:", "Reject H0" if p_value2 < alpha else "Fail to reject H0")

# Paired Sample t-test
print("  - Paired Sample t-test")
print("    - Ex: Testing if there's a significant reduction in patients' cholesterol levels after new treatment.")
print("    - H0: No significant reduction.")
print("    - H1: Significant reduction.")
print("    - p-value:", p_value3)
print("    - Result:", "Reject H0" if p_value3 < alpha else "Fail to reject H0")

# One-Sample t-test
print("  - One-Sample t-test")
print("    - Ex: Testing if average website response time is significantly less than 30 seconds.")
print("    - H0: Average response time is 30 seconds.")
print("    - H1: Average response time is different from 30 seconds.")
print("    - p-value:", p_value4)
print("    - Result:", "Reject H0" if p_value4 < alpha else "Fail to reject H0")

# Two-Sample t-test (Independent Samples)
print("  - Two-Sample t-test (Independent Samples)")
print("    - Ex: Testing if effectiveness of two advertising campaigns is different.")
print("    - H0: No difference in effectiveness.")
print("    - H1: Difference in effectiveness.")
print("    - p-value:", p_value5)
print("    - Result:", "Reject H0" if p_value5 < alpha else "Fail to reject H0")
```

- Chi-Square Test
    - Ex: Testing the independence of categorical variables.
    - H0: There's no association between gender and voting preference.
    - H1: There's an association between gender and voting preference.

```python
# Create a contingency table
# Rows: Gender (Male, Female)
# Columns: Voting Preference (A, B, C)
observed_freq = np.array([[45, 30, 25], [35, 40, 45]])

# Significance level
alpha = 0.05

# Chi-Square Test
chi2_stat, p_value, dof, expected_freq = stats.chi2_contingency(observed_freq)

# Hypotheses
hypothesis = "H0: There's no association between gender and voting preference."
hypothesis_alternative = "H1: There's an association between gender and voting preference."

# Interpretation
print("- Chi-Square Test")
print("  - Ex: Testing the independence of categorical variables.")
print("  - H0:", hypothesis)
print("  - H1:", hypothesis_alternative)
print("  - p-value:", p_value)

if p_value < alpha:
    print("  - Reject the Null Hypothesis for the Chi-Square Test.")
else:
    print("  - Fail to reject the Null Hypothesis for the Chi-Square Test.")
```

- ANOVA (Analysis of Variance)
    - Ex: Testing differences among multiple groups.
    - One-Way ANOVA
        - Ex: Comparing the average test scores of students across different teaching methods.
        - H0: The mean test scores of all teaching methods are equal.
        - H1: At least one teaching method has a different mean test score.
    - Two-Way ANOVA
        - Ex: Investigating the impact of both fertilizer type and watering frequency on plant growth.
        - H0: The effects of fertilizer type and watering frequency are independent of each other.
        - H1: There's an interaction effect between fertilizer type and watering frequency on plant growth.

```python
from sklearn.preprocessing import LabelEncoder

# Sample data for One-Way ANOVA
method1_scores = np.array([85, 88, 82, 89, 91])
method2_scores = np.array([78, 85, 80, 84, 88])
method3_scores = np.array([90, 92, 87, 91, 95])

# Significance level
alpha = 0.05

# One-Way ANOVA
f_stat, p_value = stats.f_oneway(method1_scores, method2_scores, method3_scores)

# Hypotheses
hypothesis = "H0: The mean test scores of all teaching methods are equal."
hypothesis_alternative = "H1: At least one teaching method has a different mean test score."

# Interpretation for One-Way ANOVA
print("- One-Way ANOVA")
print("  - Ex: Comparing the average test scores of students across different teaching methods.")
print("  - H0:", hypothesis)
print("  - H1:", hypothesis_alternative)
print("  - p-value:", p_value)

if p_value < alpha:
    print("  - Reject the Null Hypothesis for the One-Way ANOVA.")
else:
    print("  - Fail to reject the Null Hypothesis for the One-Way ANOVA.")
```

- Confidence Interval
    - Ex: 95% confidence interval for the population mean.
    - H0: The average response time for customer support calls is 10 minutes.
    - H1: The average response time for customer support calls is not 10 minutes.

```python
# Sample data
response_times = np.array([8.5, 10.2, 9.8, 11.0, 9.3, 10.5, 11.8, 10.9, 9.7, 10.1])

# Confidence level
confidence_level = 0.95

# Calculate the mean and standard error of the sample
sample_mean = np.mean(response_times)
sample_stddev = np.std(response_times, ddof=1)  # ddof=1 for sample standard deviation
sample_size = len(response_times)
standard_error = sample_stddev / np.sqrt(sample_size)

# Calculate the critical value for the confidence interval
alpha = 1 - confidence_level
critical_value = stats.t.ppf(1 - alpha / 2, df=sample_size - 1)

# Calculate the confidence interval
confidence_interval_lower = sample_mean - critical_value * standard_error
confidence_interval_upper = sample_mean + critical_value * standard_error

# Hypotheses
hypothesis = "H0: The average response time for customer support calls is 10 minutes."
hypothesis_alternative = "H1: The average response time for customer support calls is not 10 minutes."

# Interpretation
print("- Confidence Interval")
print("  - Ex: 95% confidence interval for the population mean.")
print("  - H0:", hypothesis)
print("  - H1:", hypothesis_alternative)
print(f"  - Sample mean: {sample_mean:.2f}")
print(f"  - Confidence Interval: ({confidence_interval_lower:.2f}, {confidence_interval_upper:.2f})")

# Check if the null hypothesis falls within the confidence interval
if confidence_interval_lower <= 10 <= confidence_interval_upper:
    print("  - Fail to reject the Null Hypothesis based on the confidence interval.")
else:
    print("  - Reject the Null Hypothesis based on the confidence interval.")
```
